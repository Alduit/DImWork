# 1.2 Работа с разряжённым форматом данных

```python
def dfToDict(df):
    S = ["site" + str(i) for i in range (1,11)]
    dict_df = {}
    for i in range(len(df)):
        d = df.iloc[i][S].value_counts().to_json()
        dict_df = d
    return dict_df
```

```python
#сохраняем тренировочную выборку
import pickle
def saveDict (dict_df, fileName):
    print("Сохраняем частотный словарь")
    with open(fileName, "wb") as fileTrainTest:
        pickle.dump(dict_df, fileTrainTest)
    print("Словарь сохранен!")
```

```python
saveDict(dfToDict(df_test),"dicTest.pkl")
saveDict(dfToDict(df_train),"dicTrain.pkl")
```

# 1.4 Конструирование новых признаков 

```python
import datetime
```

```python
def new_features(df):
    df["ym"] = df.apply(lambda x: "".join(str(x["time1"])[:str(x["time1"]).rfind("-")].split("-")), axis=1)
    
    df["time_diff"] = df.apply(lambda x: (datetime.datetime.strptime(str(x["time10"]), '%Y-%m-%d %H:%M:%S') -
 datetime.datetime.strptime(str(x["time1"]), '%Y-%m-%d %H:%M:%S')).total_seconds(), axis=1)
    
    df["max_repeat"] = df[["site" + str(i) for i in range(1, 11)]].apply(lambda x: x.value_counts().max(), axis=1)
    
    df = df.astype({"ym":"int64"})
    return df
```

```python
df_train = new_features(df_train)
df_test = new_features(df_test)
```

# Сессия 2. Построение модели машинного обучения

## 2.1 Сравнение моделей машинного обучения 

```python
X = df_train.drop(columns = ["time" +  str(i) for i in range(1, 11)], axis = 1)
X = X.drop(columns = ["target"], axis = 1)
y = df_train["target"].values
```

Раздиление данных на тестовую и обучающую выборку 

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, shuffle = True)
```

Обученм несколько моделей:
- К-ближайшего
- Случайный лес
- Гридиент бустинг

```python
from sklearn.neighbors import KNeighborsClassifier

Nieghbors = [3,5,7]
Train_score = []
Val_score = []
for nieghbors in Nieghbors:
    kn = KNeighborsClassifier(n_neighbors= nieghbors , n_jobs= -1)
    kn.fit(X_train, y_train)
    Train_score.append(kn.score(X_train,y_train))
    Val_score.append(kn.score(X_val, y_val))
```

```python
print(Train_score)
print(Val_score)
```

```python
from sklearn.metrics import classification_report

print (classification_report(y_val, kn.predict(X_val)))
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

Tree = [100,200,300]
Train_score = []
Val_score = []
for tree in Tree:
    rfc = RandomForestClassifier(n_estimators = tree, n_jobs= -1)
    rfc.fit(X_train,y_train)
    Train_score.append(kn.score(X_train,y_train))
    Val_score.append(kn.score(X_val, y_val))
```

```python
print(Train_score)
print(Val_score)
```

```python
print (classification_report(y_val, rfc.predict(X_val)))
```

```python
Tree = [300,400,500]
Train_score = []
Val_score = []
for tree in Tree:
    gbc = GradientBoostingClassifier(n_estimators= tree)
    gbc.fit(X_train,y_train)
    Train_score.append(kn.score(X_train,y_train))
    Val_score.append(kn.score(X_val, y_val))
```

```python
print(Train_score)
print(Val_score)
```

```python
print (classification_report(y_val, gbc.predict(X_val)))
```

Начинаем работать с тестовыми данным , для этого воспользуемся методом случайного леса, тк он проявил себя лучше других моделей.
